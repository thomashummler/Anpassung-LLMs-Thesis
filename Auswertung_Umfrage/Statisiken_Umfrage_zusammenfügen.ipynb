```python
dfa = pd.read_csv('/content/gdrive/MyDrive/BA-Programmcode/Auswertung_Umfrage/UserSurvey_Chatbots_A.csv')
dfb = pd.read_csv('/content/gdrive/MyDrive/BA-Programmcode/Auswertung_Umfrage/UserSurvey_Chatbots_B.csv')


print(dfa.head())
print(dfb.head())





```

```python
# Spaltenreihenfolge als Liste erhalten
cols = dfb.columns.tolist()
# Anpassung, um die ersten 4 und die letzten 2 Spalten zu überspringen
start_index = 4  # Start nach den ersten 4 Spalten
end_index = len(cols) - 2  # Endet vor den letzten 2 Spalten

for i in range(start_index, end_index, 2):  # Jedes Paar von Spalten ist zu tauschen
    cols[i], cols[i + 1] = cols[i + 1], cols[i]
# DataFrame mit der neuen Spaltenreihenfolge erstellen
df = dfb[cols]
```

```python
display(df)

```

```python
import pandas as pd

# DataFrame df ist Ihr vorhandener DataFrame

# Durchlaufen der Spaltennamen und Ersetzen von "Chatbot 1" mit "Chatbot 2" und umgekehrt
neue_spalten = []
for col in df.columns:
    if 'Chatbot 1' in col:
        neue_spalten.append(col.replace('Chatbot 1', 'Chatbot 2'))
    elif 'Chatbot 2' in col:
        neue_spalten.append(col.replace('Chatbot 2', 'Chatbot 1'))
    else:
        neue_spalten.append(col)

# Aktualisieren der Spaltennamen im DataFrame
df.columns = neue_spalten

# Überprüfen der geänderten Spaltennamen
print(df.columns)

print(df)

```

```python
# Direktes Ändern der Spaltennamen
df.columns.values[2] = 'Geben bitte Ihren Grad an Vorerfahrung mit Chatbots an  [Zeile 1]'





df_gesamt = pd.concat([dfa, df], ignore_index=True)
print(df_gesamt)



df_gesamt.to_excel('gesamt_Umfrage.xlsx', index=False)

file_path = '/content/gdrive/MyDrive/BA-Programmcode/Auswertung_Umfrage/gesamt_Umfrage.xlsx'
df_gesamt.to_excel(file_path)

```
